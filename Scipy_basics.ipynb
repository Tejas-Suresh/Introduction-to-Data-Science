```python
from scipy import stats
import numpy as np
```

```python
### Print Normal Random Variables
print(stats.norm.rvs(size = 15))
```

```python
from pylab import *

# Create some test data
dx=0.1
X  = np.arange(-2,2,dx)
Y  = exp(-X**2)

# Normalize the data to a proper PDF
Y /= (dx*Y).sum()

# Compute the CDF
CY = np.cumsum(Y*dx)

# Plot both
plot(X,Y)
plot(X,CY,'r--')

show()
```

```python
### Compute the Normal CDF of certain values.
print(stats.norm.cdf(np.array([1,-1., 0, 1, 3, 4, -2, 6])))
```
